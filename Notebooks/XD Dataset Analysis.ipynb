# **Cleaning the dataset**

## **1. Removing unnecessary and useless data fields!**

```python
# Data cleaning steps

# Remove the unwanted columns: components, fixVersions, issuetype.subtask, versions, watches.watchCount.

xd_df.drop(columns=['components', 'fixVersions', 'issuetype.subtask', 'versions', 'watches.watchCount', 'resolution.description', 'resolution.name'], inplace = True)

print("\n Columns: components, fixVersions, issuetype.subtask, versions, watches.watchCount, resolution.description, resolution.name have been removed from the dataset! \n")

print(f'\n There are now {len(xd_df.columns)} columns in the dataset.\n')
print(f"\n Printing Data Fields Below: \n {xd_df.columns} \n")
```

```python
print("Printing the new dataset: \n")

xd_df.head()
```

##**2. Removing the duplicate issues, only keeping one instance of them! Removing isses that have same issue_identifier (KEY)!**

```python
#Data cleaning steps

# 1.  Remove duplicated issue reports: it doesn't make sense to have ISSUE REPORTS that 
#     have exactly the same content since it was probably a mistake from data input.

# Select all duplicate rows based on column: key
duplicateRowsDF = xd_df[xd_df.duplicated(['key'])]
print(f"\nThere are: {len(duplicateRowsDF)} issue duplicates! \n")


print(f"{xd_df['key'].value_counts().head(20)} \n\n")
print(f"Count of column: key, before cleaning is: {xd_df['key'].count()}\n\n")
print(f"Printing some example issues that have the same identifier: \n")
print(xd_df.loc[xd_df['key'] == 'XD-1507'])
print(xd_df.loc[xd_df['key'] == 'XD-371'])
print(xd_df.loc[xd_df['key'] == 'XD-375'])
```

```python
# Continuation from previous cell

# Select all duplicate rows based on one column
#duplicateRowsDF = xd_df[xd_df.duplicated(['key'])]
#print(f"\n There are: {len(duplicateRowsDF)} issue duplicates! \n")

#print('\n*************************************************\n')

# Here I am dropping all duplicates, while only keeping the first instance!
xd_df = xd_df.drop_duplicates(subset='key', keep='first')

print(f"Count of column: key, after cleaning is: {xd_df['key'].count()}\n\n")
print(f"{xd_df['key'].value_counts()}\n\n")
print(xd_df.loc[xd_df['key'] == 'XD-1507'])
print(xd_df.loc[xd_df['key'] == 'XD-371'])
print(xd_df.loc[xd_df['key'] == 'XD-375'])


# Here I check for duplicate rows based on all columns, 
# i.e. rows where all columns have same values. 
duplicated = xd_df[xd_df.duplicated()]
print(f"\nFinally, there are: {len(duplicated)} duplicates")
```

## **3. Removing issues that do not have a creator.name recorded!**

```python
#Data cleaning steps

# 2.  Remove issue reports with the empty creator field. by default, 
#     all issues have to have a creator in JIRA

print(f"Count of column: creator.name, before cleaning is: {xd_df['creator.name'].count()}\n")
print(f"Count of missing values in the field: creator.name before cleaning, is: {xd_df['creator.name'].isnull().sum()}. \n\n\n\n")

xd_df.dropna(subset = ["creator.name"], inplace=True)

print(f"Count of column: creator.name, after cleaning is: {xd_df['creator.name'].count()}\n")
print(f"Count of missing values in the field: creator.name after cleaning, is: {xd_df['creator.name'].isnull().sum()}. \n\n")

print(xd_df['creator.name'].value_counts())

print('\n\n')
print(xd_df.loc[(xd_df['creator.name'] == '')])
print('\n\n')
xd_df.loc[(xd_df['creator.name'].isnull())]
```
